```python

import torch
import torch.nn as nn
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread("/content/car.jpg")
img = cv.resize(img, (100, 100)).astype(np.float32)

img_tensor = torch.from_numpy(img).permute(2, 0, 1).unsqueeze(0)
```

```python
conv1 = nn.Conv2d(in_channels=3, out_channels=16, kernel_size=3)

feature_map = conv1(img_tensor)
```

```python
feature_map.detach()
```

```python
plt.
```

```python
relu = nn.ReLU()
feature_map = relu(feature_map)
```

```python
pool = nn.MaxPool2d(kernel_size=2, stride=2)

feature_map = pool(feature_map)
```

```python
flattened_vec = feature_map.view(feature_map.size(0), -1)
```

```python
flattened_vec.shape
```

```python
fc = nn.Linear(in_features=38416, out_features=2)

output = fc(flattened_vec)
```
